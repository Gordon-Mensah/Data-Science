# Develop a line chart using Pandas to show how automobile sales fluctuate year by year.

```python
# Notebook 1: Task 1.1 - Line Plot for Automobile Sales

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
data = {
    'Year': np.arange(2010, 2020),
    'Car Sales': [5000, 8000, 7000, 9000, 8500, 10000, 12000, 14000, 16000, 18000],
    'Truck Sales': [5000, 7000, 5000, 9000, 7500, 10000, 10000, 11000, 11000, 12000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Line plot shows the automobile sales from year to year.
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Car Sales'], label='Car Sales', marker='o')
plt.plot(df['Year'], df['Truck Sales'], label='Truck Sales', marker='o')
plt.title("Automobile Sales from Year to Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(df['Year'])
plt.legend()
plt.grid()
plt.show()
```

# Plot different lines for categories of vehicle type and analyze the trend during recession periods.

```python
# Notebook 2: Task 1.2 - Line Plot for Vehicle Categories

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data Preparation
data = {
    'Year': np.arange(2010, 2020),
    'Car Sales': [5000, 8000, 7000, 9000, 8500, 10000, 12000, 14000, 16000, 18000],
    'Truck Sales': [5000, 7000, 5000, 9000, 7500, 10000, 10000, 11000, 11000, 12000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Line plot shows different lines for categories of vehicle type.
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.melt(id_vars='Year', value_vars=['Car Sales', 'Truck Sales']), 
             x='Year', y='value', hue='variable', marker='o')
plt.title("Sales Trends for Different Vehicle Types")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(df['Year'])
plt.grid()
plt.show()
```

# Use Seaborn to visualize and compare vehicle sales trends for recession and non-recession periods.

```python
# Notebook 3: Task 1.3 - Bar Chart for Recession vs Non-Recession

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
categories = ['Car Sales', 'Truck Sales']
values = [[3000, 4000], [10000, 12000]]  # Sample values for two periods

# Convert values to DataFrame
bar_data = pd.DataFrame(values, columns=categories, index=['Recession', 'Non-Recession'])

# Bar chart shows the vehicle sales during recession and non-recession.
bar_data.plot(kind='bar', figsize=(10, 6))
plt.title("Vehicle Sales During Recession and Non-Recession")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.legend(title="Vehicle Type")
plt.grid()
plt.show()
```

# Use subplots to compare GDP variations between recession and non-recession periods

```python
# Notebook 4: Task 1.4 - Subplot for GDP During Periods

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
data = {
    'Year': np.arange(2010, 2020),
    'GDP': [15000, 15500, 16000, 16500, 17000, 17500, 18000, 18500, 19000, 20000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Subplot showing GDP during recession and non-recession
fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# Recession Period
axes[0].plot(df['Year'], df['GDP'], marker='o', color='orange')
axes[0].set_title("GDP During Recession")
axes[0].set_ylabel("GDP")
axes[0].grid()

# Non-Recession Period
axes[1].plot(df['Year'], df['GDP'], marker='o', color='green')
axes[1].set_title("GDP During Non-Recession")
axes[1].set_ylabel("GDP")
axes[1].grid()

plt.xlabel("Year")
plt.show()
```

# Create a Bubble plot to display the impact of seasonality on automobile sales

```python
# Notebook 5: Task 1.5 - Bubble Plot for Seasonality Impact

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
data = {
    'Year': np.arange(2010, 2020),
    'Car Sales': [5000, 8000, 7000, 9000, 8500, 10000, 12000, 14000, 16000, 18000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Bubble plot showing the size variation based on seasonality
seasonality = np.random.rand(10) * 1000  # Random seasonality data
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Car Sales'], s=seasonality, alpha=0.5, c='blue', edgecolors='w', linewidth=2)
plt.title("Bubble Plot: Impact of Seasonality on Car Sales")
plt.xlabel("Year")
plt.ylabel("Car Sales")
plt.grid()
plt.show()
```

# Create a scatter plot using Matplotlib to find the correlation between average vehicle price and sales volume.

```python
# Notebook 6: Task 1.6 - Scatter Plot for Price vs Sales Volume

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
data = {
    'Year': np.arange(2010, 2020),
    'Car Sales': [5000, 8000, 7000, 9000, 8500, 10000, 12000, 14000, 16000, 18000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Scatter plot showing correlation between average vehicle price and sales volume
avg_vehicle_price = np.random.randint(15000, 30000, size=len(df))  # Sample vehicle prices
plt.figure(figsize=(10, 6))
plt.scatter(avg_vehicle_price, df['Car Sales'], color='purple', alpha=0.6)
plt.title("Scatter Plot: Average Vehicle Price vs Sales Volume")
plt.xlabel("Average Vehicle Price")
plt.ylabel("Car Sales")
plt.grid()
plt.show()
```

# Create a pie chart to display XYZAutomotives advertising expenditure during recession and non-recession periods.

```python
# Notebook 7: Task 1.7 - Pie Chart for Advertising Expenditure

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
expenditure = [500, 600]  # Sample values for recession and non-recession

# Pie chart showing advertising expenditure during recession and non-recession
plt.figure(figsize=(8, 8))
plt.pie(expenditure, labels=['Recession', 'Non-Recession'], autopct='%1.1f%%', startangle=140)
plt.title("Proportional Advertising Expenditure During Recession and Non-Recession")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# Develop a pie chart to display advertisement expenditure for each vehicle type during the recession period.

# Develop a line plot to analyze the effect of unemployment on vehicle sales during the recession period.

```python
# Notebook 9: Task 1.9 - Line Plot for Unemployment Rate

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample Data Preparation
data = {
    'Year': np.arange(2010, 2020),
    'Unemployment Rate': [8, 9, 10, 12, 10, 8, 7, 6, 5, 4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Line plot showing unemployment rate and its effect on vehicle sales
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Unemployment Rate'], label='Unemployment Rate', marker='o', color='red')
plt.title("Unemployment Rate and Vehicle Sales Over Time")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(df['Year'])
plt.grid()
plt.legend()
plt.show()
```
