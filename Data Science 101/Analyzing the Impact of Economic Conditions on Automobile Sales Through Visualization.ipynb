# Develop a line chart using Pandas to show how automobile sales fluctuate year by year.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Year': [2010,2011,2012,2013,2014],
    'Sales' : [10000, 15000, 12000, 18000, 16000],
})

#John Jerry Gordon-Mensah
df.plot(x='Year',y='Sales', kind = 'line')
plt.title("Automobile Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

# Plot different lines for categories of vehicle type and analyze the trend during recession periods.

```python
# Sample data for vehicle types
df = pd.DataFrame({
    'Year': [2010, 2010, 2011, 2011, 2012, 2012, 2013, 2013, 2014, 2014],
    'Car Sales': [5000, 7000, 8000, 9000, 7000, 6500, 9000, 8000, 8500, 7700],
    'Vehicle Type': ['Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck', 'Car', 'Truck'],
    'Truck Sales': [4000, 2000, 3000 , 1500,  5000, 4000, 2500, 3000, 2500, 4700]
})

df.plot(x='Year', y=['Car Sales',  'Truck Sales'], kind='line')
plt.title("Sales Trends for Car vs Truck During Recession")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

# Use Seaborn to visualize and compare vehicle sales trends for recession and non-recession periods.

```python

sns.lineplot(data=df, x='Year', y='Car Sales', hue='Vehicle Type')
plt.title("Sales Comparison: Recession vs Non-recession")
plt.show()
```

# Use subplots to compare GDP variations between recession and non-recession periods

```python
df_recession = pd.DataFrame({
    'Year': [2008, 2009, 2010, 2011],
    'GDP': [-2.5, -0.1, 1.5, 2.0]
})

# Sample data for non-recession period
df_non_recession = pd.DataFrame({
    'Year': [2012, 2013, 2014, 2015],
    'GDP': [2.5, 3.0, 3.2, 3.5]
})
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Recession period
axes[0].plot(df_recession['Year'], df_recession['GDP'])
axes[0].set_title("GDP During Recession")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("GDP Growth (%)")

# Non-recession period
axes[1].plot(df_non_recession['Year'], df_non_recession['GDP'])
axes[1].set_title("GDP During Non-Recession")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("GDP Growth (%)")

plt.show()
```

# Create a Bubble plot to display the impact of seasonality on automobile sales

```python
data = {'Date': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-04-05'],
        'Sales': [200, 300, 150, 400],
        'Seasonality': [1.2, 1.5, 1.1, 1.3]}

df = pd.DataFrame(data)

# Convert 'Date' column to datetime and extract the month
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Bubble plot with month as the x-axis
plt.scatter(df['Month'], df['Sales'], s=df['Seasonality']*100)
plt.title("Impact of Seasonality on Automobile Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Create a scatter plot using Matplotlib to find the correlation between average vehicle price and sales volume.

```python
data = {'Sales': [200, 300, 150, 400], 'Seasonality': [1.2, 1.5, 1.1, 1.3]}
df = pd.DataFrame(data)

# Manually adding the 'Average Price' column
df['Average Price'] = [25000, 27000, 22000, 26000]

# Scatter plot of 'Average Price' vs 'Sales'
plt.scatter(df['Average Price'], df['Sales'])
plt.title("Correlation Between Average Price and Sales Volume")
plt.xlabel("Average Price")
plt.ylabel("Sales")
plt.show()
```

# Develop a line plot to analyze the effect of unemployment on vehicle sales during the recession period.

```python
data = {
    'Year': [2008, 2009, 2010, 2011, 2012, 2013, 2014],
    'Unemployment Rate': [7.8, 9.3, 9.6, 8.9, 8.1, 7.4, 6.2],  # Example unemployment rates
    'Sales': [15000, 12000, 10000, 13000, 14000, 16000, 17000]  # Example sales numbers
}

df_recession = pd.DataFrame(data)


print("DataFrame:")
print(df_recession)


plt.figure(figsize=(10, 5))
plt.plot(df_recession['Year'], df_recession['Unemployment Rate'], marker='o', label='Unemployment Rate', color='orange')
plt.ylabel('Unemployment Rate (%)', color='orange')
plt.xlabel('Year')
plt.twinx()  # Create a second y-axis for sales
plt.plot(df_recession['Year'], df_recession['Sales'], marker='o', label='Vehicle Sales', color='blue')
plt.ylabel('Vehicle Sales', color='blue')


plt.title("Unemployment Rate and Vehicle Sales During Recession")
plt.grid()
plt.xticks(df_recession['Year'])  # Show all years on the x-axis
plt.legend(loc='upper left')

plt.show()
```
